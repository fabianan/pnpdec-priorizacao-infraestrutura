## 2. Carregamento e limpeza do Relatório Gerencial

```python
df_rel = pd.read_excel(RELATORIO_XLSX, sheet_name=ABA_RELATORIO,
                        dtype=str, engine='openpyxl')

# Padronizar nomes de colunas (aplicado aqui, como feito em qKOqEMPImGOz)
df_rel.columns = (
    df_rel.columns
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('º', 'num')
    .str.replace('ç', 'c')
)

# ── Normaliza código IBGE para 7 dígitos (padrão do pipeline base) ───────
df_rel[COL_IBGE] = (
    df_rel[COL_IBGE]
    .astype(str)
    .str.strip()
    .str.extract(r'(\d+)')[0]          # extrai só os dígitos
    .str.zfill(7)                       # padroniza para 7 dígitos
)

# ── Normaliza COBRADE (remove pontos/hifens, mantém string) ──────────────
df_rel[COL_COBRADE] = (
    df_rel[COL_COBRADE]
    .astype(str)
    .str.strip()
    .str.replace(r'[.\-]', '', regex=True)
)

# ── Converte valor para numérico ──────────────────────────────────────────
if COL_VALOR in df_rel.columns:
    df_rel[COL_VALOR] = (
        df_rel[COL_VALOR]
        .astype(str)
        .str.replace(r'[R$\s.]', '', regex=True)
        .str.replace(',', '.', regex=False)
        .pipe(pd.to_numeric, errors='coerce')
    )

print(f'Relatório carregado: {len(df_rel):,} registros')
print(f'Municípios únicos  : {df_rel[COL_IBGE].nunique():,}')
if COL_ANO in df_rel.columns:
    anos = df_rel[COL_ANO].unique()
    print(f'Período coberto    : {sorted(anos)}')
```

## 3. Carregamento e filtragem COBRADE

Filtramos apenas os desastres **aderentes ao tema de infraestrutura urbana preventiva**:  
inundações, enxurradas, alagamentos e deslizamentos de terra.  
São os eventos que drenagem, saneamento e contenção de encostas poderiam mitigar.

```python
df_cobrade = pd.read_csv(COBRADE_CSV, dtype=str)
print('Colunas COBRADE:', df_cobrade.columns.tolist())
print(df_cobrade.head(5).to_string())
```

```python
# ── Diagnóstico: o que existe no relatório? ───────────────────────────────
print("=== COBRADE únicos no relatório (primeiros 40) ===")
cobrades_relatorio = df_rel[COL_COBRADE].dropna().unique()
cobrades_relatorio.sort()
for c in cobrades_relatorio[:40]:
    print(f"  '{c}'  (len={len(str(c))})")

print(f"\nTotal de COBRADEs únicos no relatório: {len(cobrades_relatorio)}")

print("\n=== COBRADE_NUM únicos no arquivo cobrade.csv (primeiros 40) ===")
for c in sorted(df_cobrade['COBRADE_NUM'].dropna().unique())[:40]:
    print(f"  '{c}'")
```

```python
# ── Filtragem por COBRADE_NUM do arquivo de referência ───────────────────
# Abordagem robusta: lê os tipos aderentes direto do cobrade.csv
# em vez de manter dicionário fixo no código

# O tema do trabalho é infraestrutura urbana preventiva: drenagem, saneamento e contenção de encostas. Logo:

TIPOS_ADERENTES = {
    # Movimento de massa (deslizamentos) — SUBGRUPO 1.1.3
    'Movimento de massa': 'Deslizamento',
    # Hidrológico — SUBGRUPO 1.3.1 inundações, 1.3.2 enxurradas, 1.3.3 alagamentos
    'Inundação': 'Inundação',
    'Enxurrada': 'Enxurrada',
    'Alagamento': 'Alagamento',
}

# Filtra o cobrade.csv pelos tipos aderentes pelo campo TIPO
mask_cobrade_aderente = (
    df_cobrade['TIPO'].str.contains('Movimento de massa|Inundação|Enxurrada|Alagamento',
                                     case=False, na=False)
    | df_cobrade['SUBGRUPO'].str.contains('Hidrológico', case=False, na=False)
)

df_cobrade_aderente = df_cobrade[mask_cobrade_aderente].copy()
print(f"COBRADEs aderentes encontrados no arquivo de referência: {len(df_cobrade_aderente)}")
print(df_cobrade_aderente[['COBRADE_NUM','SUBGRUPO','TIPO','SUBTIPO']].to_string())

# Extrai os COBRADE_NUM aderentes como conjunto de busca
cobrades_num_aderentes = set(df_cobrade_aderente['COBRADE_NUM'].str.strip())

# Classifica cada linha do relatório

def classificar_tipo(cobrade_rel):
    row = df_cobrade_aderente[df_cobrade_aderente['COBRADE_NUM'] == cobrade_rel]
    if len(row) == 0:
        return None
    tipo = row['TIPO'].values[0]
    subtipo_val = row['SUBGRUPO'].values[0]
    # Usa SUBGRUPO para hidrológicos (mais confiável que TIPO nesse arquivo)
    if 'Hidrológico' in subtipo_val:
        if '1. Inundações' in tipo or 'Inundação' in tipo:
            return 'Inundação'
        if 'Enxurrada' in tipo:
            return 'Enxurrada'
        if 'Alagamento' in tipo:
            return 'Alagamento'
        return 'Hidrológico outro'
    if 'Movimento de massa' in tipo:
        return 'Deslizamento'
    return 'Outro aderente'

# Normaliza o COBRADE do relatório para 5 dígitos (remove pontos/espaços)
df_rel['cobrade_norm'] = (
    df_rel[COL_COBRADE]
    .astype(str)
    .str.strip()
    .str.replace(r'[.\-\s]', '', regex=True)
    .str.zfill(5)
)

mask_aderente = df_rel['cobrade_norm'].isin(cobrades_num_aderentes)
df_rel['tipo_desastre'] = df_rel['cobrade_norm'].apply(
    lambda x: classificar_tipo(x) if x in cobrades_num_aderentes else None
)

df_aderente = df_rel[mask_aderente].copy()

print(f'\nRegistros totais no relatório  : {len(df_rel):,}')
print(f'Registros aderentes ao tema    : {len(df_aderente):,}')
print(f'Municípios com evento aderente : {df_aderente[COL_IBGE].nunique():,}')
print(f'\nDistribuição por tipo:')
print(df_aderente['tipo_desastre'].value_counts().to_string())
```

## 4. Agregação por município e cruzamento com o ranking

```python
# ── Agrega eventos aderentes por município ────────────────────────────────
agg_dict = {
    COL_MUNICIPIO: 'first',
    COL_COBRADE: 'count',      # total de eventos aderentes
}
if COL_VALOR in df_aderente.columns:
    agg_dict[COL_VALOR] = 'sum'

df_eventos = (
    df_aderente
    .groupby(COL_IBGE)
    .agg(agg_dict)
    .reset_index()
    .rename(columns={
        COL_IBGE    : 'cod_ibge',
        COL_MUNICIPIO : 'municipio',
        COL_COBRADE : 'n_eventos_aderentes',
        COL_VALOR   : 'valor_total_liberado',
    })
)

print(f'Municípios com ao menos 1 evento aderente: {len(df_eventos):,}')
print(df_eventos.describe().to_string())
```

```python
# ── Carrega ranking do pipeline base ─────────────────────────────────────
df_rank = pd.read_csv(PRIORIZADOS_CSV, dtype={'cod_ibge': str})
df_rank['cod_ibge'] = df_rank['cod_ibge'].str.zfill(7)

print(f'Municípios no ranking: {len(df_rank):,}')
print(f'Colunas disponíveis  : {df_rank.columns.tolist()}')
```

```python
# ── Enriquecimento com nomes de municípios via API IBGE ───────────────────
# Feito aqui no notebook de validação para não depender do CSV do notebook 1
print('Buscando nomes de municípios na API IBGE...')
try:
    resp = requests.get(
        'https://servicodados.ibge.gov.br/api/v1/localidades/municipios?view=nivelado',
        timeout=30
    )
    resp.raise_for_status()
    df_ibge = pd.DataFrame(resp.json())
    df_ibge['cod_ibge'] = df_ibge['municipio-id'].astype(str).str.zfill(7)
    df_ibge = df_ibge[['cod_ibge', 'municipio-nome', 'UF-sigla']].rename(
        columns={'municipio-nome': 'nome_municipio', 'UF-sigla': 'uf_ibge'}
    )
    # Merge — nome_municipio do IBGE substitui qualquer versão parcial do CSV
    df_rank = df_rank.drop(columns=['nome_municipio', 'uf'], errors='ignore')
    df_rank = df_rank.merge(df_ibge, on='cod_ibge', how='left')
    # Usa uf_ibge como coluna 'uf' padrão
    if 'uf' not in df_rank.columns:
        df_rank = df_rank.rename(columns={'uf_ibge': 'uf'})
    n_sem_nome = df_rank['nome_municipio'].isna().sum()
    print(f'[OK] {len(df_ibge):,} municípios carregados do IBGE')
    print(f'     Sem nome após merge: {n_sem_nome} (cod_ibge não encontrado na API)')
except Exception as e:
    print(f'[AVISO] API IBGE indisponível: {e}')
    print('        Usando cod_ibge como fallback nos rótulos')
    if 'nome_municipio' not in df_rank.columns:
        df_rank['nome_municipio'] = df_rank['cod_ibge']
```

```python
# ── Cruzamento ranking × histórico de desastres ──────────────────────────
df = df_rank.merge(df_eventos, on='cod_ibge', how='left')

# Flag: município tem histórico de desastres aderentes?
df['tem_historico'] = df['n_eventos_aderentes'].notna() & (df['n_eventos_aderentes'] > 0)
df['n_eventos_aderentes'] = df['n_eventos_aderentes'].fillna(0).astype(int)

print(f'Municípios no cruzamento        : {len(df):,}')
print(f'Com histórico aderente          : {df["tem_historico"].sum():,}')
print(f'Sem histórico (invisíveis)      : {(~df["tem_historico"]).sum():,}')
print()
print('Distribuição por faixa:')
print(
    df.groupby('faixa_prioridade')['tem_historico']
    .agg(['sum', 'count'])
    .rename(columns={'sum': 'com_historico', 'count': 'total'})
    .assign(pct=lambda x: (x.com_historico / x.total * 100).round(1))
    .to_string()
)
```

## 5. Análise 1 — Correlação entre score e frequência de desastres

```python
# ── Correlação de Spearman ────────────────────────
# Usamos Spearman porque n_eventos tem distribuição muito assimétrica

rho, pval = stats.spearmanr(
    df['score_prioridade'],
    df['n_eventos_aderentes']
)
print(f'Correlação de Spearman (score × n_eventos): ρ = {rho:.3f}  (p = {pval:.4f})')
if pval < 0.05:
    direcao = 'positiva' if rho > 0 else 'negativa'
    print(f'→ Correlação {direcao} estatisticamente significativa.')
    if rho > 0:
        print('  Municípios mais vulneráveis tendem a ter MAIS eventos — modelo tem respaldo empírico.')
    else:
        print('  Municípios mais vulneráveis têm MENOS eventos — possível subnotificação ou invisibilidade.')
else:
    print('→ Sem correlação significativa. Interpretar com cautela (ver análise de invisíveis).')

# Por faixa
print('\nMédia de eventos por faixa:')
print(
    df.groupby('faixa_prioridade')['n_eventos_aderentes']
    .agg(['mean', 'median', 'max'])
    .round(2)
    .to_string()
)
```

## 6. Análise 2 — Municípios "invisíveis"

Municípios com score ALTA ou MEDIA mas **sem nenhum registro** no  
relatório gerencial. Isso pode indicar:
- Ausência real de desastre no período coberto pelo relatório
- Incapacidade institucional de solicitar recursos (não aparecem no sistema)
- Subnotificação (desastre ocorreu mas não foi registrado no S2iD)

```python
# ── Municípios invisíveis: score >= MEDIA e sem histórico ────────────────
df_invis = df[
    df['faixa_prioridade'].isin(['CRITICA', 'ALTA', 'MEDIA']) &
    (~df['tem_historico'])
].copy()

print(f'Total de municípios prioritários (ALTA/MEDIA): '
      f'{len(df[df["faixa_prioridade"].isin(["CRITICA","ALTA","MEDIA"])]):,}')
print(f'Desses, SEM histórico de desastre aderente    : {len(df_invis):,}')
print(f'  → {len(df_invis)/len(df[df["faixa_prioridade"].isin(["CRITICA","ALTA","MEDIA"])])*100:.1f}%'
      f' são "invisíveis" para a resposta federal')

print(f'\nDistribuição dos invisíveis por faixa:')
print(df_invis['faixa_prioridade'].value_counts().to_string())

if 'uf' in df_invis.columns:
    print(f'\nTop UF com mais invisíveis:')
    print(df_invis['uf'].value_counts().head(10).to_string())
```

```python
# ── Tabela: Top 20 invisíveis por score ───────────────────────────────────
cols_exibir = ['nome_municipio', 'uf', 'score_prioridade',
               'ivs_infraestrutura_urbana', 'faixa_prioridade']
cols_ok = [c for c in cols_exibir if c in df_invis.columns]

top_invis = df_invis.sort_values('score_prioridade', ascending=False).head(20)
print('Top 20 municípios invisíveis (maior score sem histórico de desastre):')
print(top_invis[cols_ok].to_string(index=False))

# Salva
df_invis.sort_values('score_prioridade', ascending=False).to_csv(
    'data/processed/municipios_invisiveis.csv', index=False
)
print('\n[OK] Lista salva em data/processed/municipios_invisiveis.csv')
```

```python
# ── Gráfico: Comparação invisíveis vs detectados ──────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))

por_faixa = (
    df[df['faixa_prioridade'].isin(['CRITICA', 'ALTA', 'MEDIA'])]
    .groupby('faixa_prioridade')['tem_historico']
    .value_counts(normalize=True)
    .mul(100)
    .rename('pct')
    .reset_index()
)

faixas = [f for f in ['CRITICA', 'ALTA', 'MEDIA'] if f in por_faixa['faixa_prioridade'].unique()]
x = np.arange(len(faixas))
w = 0.35

for i, (historico, label, cor) in enumerate([
    (True, 'Com histórico de desastre', '#1D9E75'),
    (False, 'Sem histórico (invisíveis)', '#E8593C')
]):
    vals = []
    for f in faixas:
        sub = por_faixa[(por_faixa['faixa_prioridade'] == f) &
                         (por_faixa['tem_historico'] == historico)]
        vals.append(sub['pct'].values[0] if len(sub) > 0 else 0)
    bars = ax.bar(x + i*w - w/2, vals, w, label=label, color=cor, alpha=0.85)
    ax.bar_label(bars, fmt='%.0f%%', padding=3, fontsize=9)

ax.set_xticks(x)
ax.set_xticklabels(faixas)
ax.set_ylabel('% dos municípios na faixa')
ax.set_title('Municípios com e sem histórico de desastre aderente\npor faixa de prioridade')
ax.legend()
ax.set_ylim(0, 115)

plt.tight_layout()
# Salva local (temporário na VM) e no Drive (persistente)
nome_arquivo = 'graf2_invisiveis.png'
for dest in [LOCAL_GRAF / nome_arquivo,
             DRIVE_GRAF  / nome_arquivo]:
    plt.savefig(dest, bbox_inches='tight')
plt.show()
print(f'[OK] {nome_arquivo} salvo em Drive e local')

plt.show()
print('[OK] Gráfico 2 salvo em data/processed/graf2_invisiveis.png')
```

## 7. Análise 3 — Score ampliado com componente de exposição real

Incorpora a frequência histórica de desastres aderentes como  
**terceiro componente** do score. Isso resolve a faixa CRÍTICA vazia  
do pipeline base, que dependia apenas de IVS e CadÚnico.  

**Ponderação proposta:**
- IVS infraestrutura urbana: 45% (era 55%)
- Vulnerabilidade sanitária CadÚnico: 35% (era 45%)
- Exposição histórica a desastres aderentes: 20% (novo)

```python
# ── Normaliza n_eventos para escala 0-100 ─────────────────────────────────
max_eventos = df['n_eventos_aderentes'].max()
if max_eventos > 0:
    df['exposicao_norm'] = (df['n_eventos_aderentes'] / max_eventos * 100).round(2)
else:
    df['exposicao_norm'] = 0
    print('[AVISO] Nenhum evento encontrado — verifique o cruzamento COBRADE.')

# ── Verifica que o score original existe e extrai seus componentes ─────────
# O score base = 55*ivs_norm + 45*saneamento_norm
# Recalculamos com nova ponderação

# Normaliza IVS (já está em 0-1 no pipeline base)
df['ivs_norm_100'] = (df['ivs_infraestrutura_urbana'].astype(float) * 100).round(2)

# Saneamento: derivado da renda (proxied do pipeline base)
# Inverso da renda normalizada (menor renda = maior vulnerabilidade)
renda_max = df['renda_media_pc'].astype(float).max()
df['saneamento_norm'] = ((1 - df['renda_media_pc'].astype(float) / renda_max) * 100).round(2)

# ── Score ampliado ─────────────────────────────────────────────────────────
df['score_ampliado'] = (
    0.45 * df['ivs_norm_100'] +
    0.35 * df['saneamento_norm'] +
    0.20 * df['exposicao_norm']
).round(2)

# ── Nova classificação por faixa ──────────────────────────────────────────
def classificar_faixa(score):
    if score > 75:  return 'CRITICA'
    if score > 50:  return 'ALTA'
    if score > 25:  return 'MEDIA'
    return 'BAIXA'

df['faixa_ampliada'] = df['score_ampliado'].apply(classificar_faixa)

print('Distribuição original vs ampliada:')
comp = pd.DataFrame({
    'Original' : df['faixa_prioridade'].value_counts(),
    'Ampliado' : df['faixa_ampliada'].value_counts(),
}).fillna(0).astype(int)
print(comp.to_string())
```

```python
# ── Top 20 no score ampliado ──────────────────────────────────────────────
cols_amp = ['município', 'uf', 'score_prioridade', 'score_ampliado',
            'faixa_prioridade', 'faixa_ampliada', 'n_eventos_aderentes',
            'ivs_infraestrutura_urbana']
cols_ok = [c for c in cols_amp if c in df.columns]

top_amp = df.sort_values('score_ampliado', ascending=False).head(20)
print('Top 20 municípios — score ampliado (IVS + CadÚnico + Exposição histórica):')
print(top_amp[cols_ok].to_string(index=False))

# Salva resultado completo
df.sort_values('score_ampliado', ascending=False).to_csv(
    'data/processed/municipios_score_ampliado.csv', index=False
)
print('\n[OK] Score ampliado salvo em data/processed/municipios_score_ampliado.csv')
```

```python
# ── Gráfico 3: score original vs ampliado (top 30) ───────────────────────
top30 = df.sort_values('score_ampliado', ascending=False).head(30).copy()

# Rótulo robusto: nome do IBGE → fallback para cod_ibge se ainda NaN
top30['rotulo'] = (
    top30['nome_municipio']
    .fillna('')
    .str.strip()
    .replace('', None)                        # string vazia vira NaN
    .fillna(top30['cod_ibge'])                # fallback para código
)

# Adiciona UF ao rótulo para desambiguar municípios homônimos
if 'uf' in top30.columns:
    top30['rotulo'] = top30['rotulo'] + ' (' + top30['uf'].fillna('') + ')'

fig, ax = plt.subplots(figsize=(11, 9))
y = np.arange(len(top30))

ax.barh(y - 0.2, top30['score_prioridade'], 0.35,
        label='Score original (IVS + CadÚnico)', color='#378ADD', alpha=0.85)
ax.barh(y + 0.2, top30['score_ampliado'], 0.35,
        label='Score ampliado (+ Exposição histórica)', color='#E8593C', alpha=0.85)

ax.set_yticks(y)
ax.set_yticklabels(top30['rotulo'].values, fontsize=8)
ax.set_xlabel('Score de priorização')
ax.set_title(
    'Comparativo: score original vs score ampliado\n'
    '(Top 30 municípios por score ampliado)',
    pad=12
)
ax.legend()
ax.invert_yaxis()

plt.tight_layout()
nome_arquivo = 'graf3_score_ampliado.png'
for dest in [LOCAL_GRAF / nome_arquivo,
             DRIVE_GRAF  / nome_arquivo]:
    plt.savefig(dest, bbox_inches='tight')
plt.show()
print('[OK] graf3_score_ampliado.png salvo em local e Drive')
```
